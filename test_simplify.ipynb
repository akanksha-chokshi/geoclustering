```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Function to simplify geometries
def simplify_geometries(gdf, tolerance=0.001):
    gdf['geometry'] = gdf['geometry'].simplify(tolerance, preserve_topology=True)
    return gdf

# Load the GeoJSON file
geojson_file = 'clusters_geojson/cluster_18.geojson'  # Replace with your GeoJSON file path
gdf = gpd.read_file(geojson_file)

# Simplify the geometries
simplified_gdf = simplify_geometries(gdf.copy(), tolerance=0.001)

# Save the simplified geometries to a new GeoJSON file
simplified_geojson_file = 'simplified_cluster_18.geojson'
simplified_gdf.to_file(simplified_geojson_file, driver='GeoJSON')

# Plotting the original and simplified geometries
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

# Plot original geometries
gdf.boundary.plot(ax=ax[0], color='blue', edgecolor='black')
ax[0].set_title('Original Geometries')

# Plot simplified geometries
simplified_gdf.boundary.plot(ax=ax[1], color='red', edgecolor='black')
ax[1].set_title('Simplified Geometries (Tolerance = 0.001)')

plt.tight_layout()
plt.show()

print(f"Simplified GeoJSON file saved as {simplified_geojson_file}")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Function to simplify geometries
def simplify_geometries(gdf, tolerance=0.001):
    gdf['geometry'] = gdf['geometry'].simplify(tolerance, preserve_topology=True)
    return gdf

# Load the GeoJSON file
geojson_file = 'clusters_geojson/cluster_18.geojson'  # Replace with your GeoJSON file path
gdf = gpd.read_file(geojson_file)

# Simplify the geometries
simplified_gdf = simplify_geometries(gdf.copy(), tolerance=0.001)

# Save the simplified geometries to a new GeoJSON file
simplified_geojson_file = 'simplified_cluster_18.geojson'
simplified_gdf.to_file(simplified_geojson_file, driver='GeoJSON')

# Plotting the original and simplified geometries
fig, ax = plt.subplots(1, 2, figsize=(20, 10))

# Plot original geometries
for geom in gdf.geometry:
    if geom.geom_type == 'Polygon' or geom.geom_type == 'MultiPolygon':
        gdf.plot(ax=ax[0], edgecolor='black')
ax[0].set_title('Original Geometries')

# Plot simplified geometries
for geom in simplified_gdf.geometry:
    if geom.geom_type == 'Polygon' or geom.geom_type == 'MultiPolygon':
        simplified_gdf.plot(ax=ax[1], edgecolor='black')
ax[1].set_title('Simplified Geometries (Tolerance = 0.001)')

plt.tight_layout()
plt.show()

print(f"Simplified GeoJSON file saved as {simplified_geojson_file}")
```
